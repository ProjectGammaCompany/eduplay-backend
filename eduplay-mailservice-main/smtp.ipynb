```python
from smtplib import SMTP
import ssl
import certifi
```

```python
srv = SMTP(host="smtp.hse.ru", port=587)
```

```python
context = ssl.create_default_context()
context.load_verify_locations(certifi.where())
# srv.ehlo()  # Identify yourself to the server
srv.starttls(context=context) # Secure the connection
# srv.ehlo()  # Re-identify yourself over the secure connection
srv.login("aignatov@hse.ru", "hseWhyNotStudy114&81")
```

```python
srv.quit()
```
